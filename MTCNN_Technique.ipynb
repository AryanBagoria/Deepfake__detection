```python
import cv2
from mtcnn import MTCNN
from matplotlib import pyplot as plt
```

Filtering detections with confidence greater than the confidence threshold and plotting detections

# Initialize the MTCNN Detector

```python
detector = MTCNN()
```

 # Load an Image

```python
image_path = "17366.jpg"
image = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
```

 # Detect Faces

```python
result = detector.detect_faces(image)
```

# Analyze the Result

```python
if len(result) == 0:
    print("No faces detected.")
else:
    print(f"Number of faces detected: {len(result)}")
    for i, det in enumerate(result):
        # Print detection data
        print(f"Face {i+1}:")
        print(f"Confidence: {det['confidence']}")
        print(f"Bounding box: {det['box']}")
        print(f"Keypoints: {det['keypoints']}")

        # Draw bounding box and keypoints
        x, y, width, height = det['box']
        cv2.rectangle(image, (x, y), (x+width, y+height), (0, 155, 255), 2)
        for key, point in det['keypoints'].items():
            cv2.circle(image, point, 2, (0, 155, 255), 2)
```

# Display image 

```python
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis('on')
plt.show()
```
